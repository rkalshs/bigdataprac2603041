```python
vc = df[df.supp == "VC"][['len', 'dose']]
oj = df[df.supp == "OJ"][['len', 'dose']]
```

```python
plt.plot(vc.dose, vc.len, color='cyan', marker='o', label="아스코르브산(VC)", lw = 0.7)
plt.plot(oj.dose, oj.len, color='deeppink', marker='^', ls=':', label="오렌지 주스(OJ)", lw = 0.7, alpha = 0.5)
plt.title("기니피그의 비타민과 치아성장 관계")
plt.xlabel("비타민 용량")
plt.ylabel("치아성장 세포길이")
plt.legend();
```
